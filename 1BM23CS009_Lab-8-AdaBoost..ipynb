```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# load data
df = pd.read_csv('/content/income (1).csv')

# encode
df = df.apply(LabelEncoder().fit_transform)

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# split
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.2)

# base model
m = AdaBoostClassifier(n_estimators=10)
m.fit(Xtr, ytr)
p = m.predict(Xte)

print("Acc(10):", accuracy_score(yte, p))

# tuning
best = 0
best_n = 0

for i in [10,20,50,100]:
    m = AdaBoostClassifier(n_estimators=i)
    m.fit(Xtr, ytr)
    p = m.predict(Xte)
    a = accuracy_score(yte, p)
    if a > best:
        best, best_n = a, i

print("Best Acc:", best)
print("Best Trees:", best_n)

# confusion matrix
print("CM:\n", confusion_matrix(yte, p))
```
